```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import random

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_results_complete_df = pd.merge(study_results, mouse_metadata, how="left",on = ["Mouse ID","Mouse ID"])

# Display the data table for preview
study_results_complete_df.head()
```

```python
# Checking the number of mice.
mice_total = len(study_results_complete_df["Mouse ID"].unique())
mice_total
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.

#extra duplicates df here
duplicates_df = study_results_complete_df[study_results_complete_df.duplicated(['Mouse ID','Timepoint'])]
#unique mouse ID's of duplicates.  (there's just one.  it's g989)
mouse_id_duplicates = duplicates_df['Mouse ID'].unique()
mouse_id_duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mice by their ID.

# there's only one mouse that needs to be purged, but just incase, there's more later, purging all possible mouse id's with duplicates
for mouse in mouse_id_duplicates:
    study_results_complete_df = study_results_complete_df.loc[study_results_complete_df['Mouse ID']!=mouse]
study_results_complete_df.head(5)
```

```python
# Checking the number of mice in the clean DataFrame.
mice_total = len(study_results_complete_df["Mouse ID"].unique())
mice_total
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

tumor_mean = study_results_complete_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
tumor_median = study_results_complete_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
tumor_variance = study_results_complete_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
tumor_stdev = study_results_complete_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
tumor_sem = study_results_complete_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.

drug_tumor_stats_df = pd.DataFrame({
    "Mean Tumor Volume":tumor_mean,
    "Median Tumor Volume":tumor_median,
    "Tumor Volume Variance":tumor_variance,
    "Tumor Volume Std. Dev.":tumor_stdev,
    "Tumor Volume Std. Err.":tumor_sem
})

drug_tumor_stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
drug_tumor_stats_df = study_results_complete_df.groupby('Drug Regimen')["Tumor Volume (mm3)"].agg(['mean','median','var','std','sem'])
drug_tumor_stats_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
mouse_final_timepoint = study_results_complete_df.groupby('Mouse ID')['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

#turning mouse_final_timepoint into a dataframe for ease of coding
mouse_final_timepoint = pd.DataFrame(mouse_final_timepoint)

#filling in info using merge for only final timepoints
mouse_final_timepoint = pd.merge(mouse_final_timepoint, study_results_complete_df, how='left', on=['Mouse ID','Timepoint'])

#filtering for just the 4 treatments and resetting index
mouse_final_timepoint = mouse_final_timepoint.loc[
    (mouse_final_timepoint['Drug Regimen']=='Capomulin') |
    (mouse_final_timepoint['Drug Regimen']=='Ramicane') |
    (mouse_final_timepoint['Drug Regimen']=='Infubinol') |
    (mouse_final_timepoint['Drug Regimen']=='Ceftamin')
].reset_index(drop=True)
mouse_final_timepoint.head()
```

```python
#determining outliers

#initiating variables
drugs = ['Capomulin',
         'Ramicane',
         'Infubinol',
         'Ceftamin']
tumor_vol = []
outliers = []

for drug in drugs:
    #creating a list of all tumor volumes for specific drug
    tumor_vol_df = mouse_final_timepoint['Tumor Volume (mm3)'][(mouse_final_timepoint['Drug Regimen']==drug)]
    tumor_vol = tumor_vol_df.values
    
    #calculating upper and lower bound for outliers for tumor volumes
    quartiles = tumor_vol_df.quantile([0.25,0.5,0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - 1.5*iqr
    upper_bound = upperq + 1.5*iqr
    
    #calculate outliers
    outliers = [vol for vol in tumor_vol if vol < lower_bound or vol > upper_bound]
    
    #printing outliers
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

#intitializing a list to hold lists.  this will be used for boxplot
drug_tumor_volumes = []

#iterating through relevant drugs
for drug in drugs:
    
    #creating a list of all tumor volumes for specific drug
    tumor_vol_df = mouse_final_timepoint['Tumor Volume (mm3)'][(mouse_final_timepoint['Drug Regimen']==drug)]
    tumor_vol = tumor_vol_df.values
    
    # Append the tumor volumes to the list of lists
    drug_tumor_volumes.append(tumor_vol)

#boxplot using drug_tumor_volumes
plt.boxplot(drug_tumor_volumes,flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 12})
plt.ylabel('Final Tumor Volumes (mm3)')
plt.xticks(range(1,5),drugs)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

#generating all unique mice that were dosed with Capomulin
study_results_capomulin_df = study_results_complete_df[(study_results_complete_df['Drug Regimen']=='Capomulin')]
capomulin_mice = study_results_capomulin_df['Mouse ID'].unique()

#generating a random mouse on Capomulin
random_number = random.randint(0,len(capomulin_mice)-1)
random_mouse = capomulin_mice[random_number]

#generating df of capomulin treatments for random mouse
capomulin_treatment_df = study_results_complete_df[(study_results_complete_df['Mouse ID']==random_mouse)]

#Creating a plot for random mouse on Capomulin.  tumor volume vs timepoint
x_values = capomulin_treatment_df['Timepoint']
y_values = capomulin_treatment_df['Tumor Volume (mm3)']
plt.plot(x_values, y_values, label = random_mouse)
plt.title("Capomulin treatment on mouse " + random_mouse)
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

#get average tumor volume (throughout the experiment)
average_volume_vs_weight = study_results_capomulin_df.groupby(['Mouse ID']).agg({
    'Tumor Volume (mm3)': 'mean',
    'Weight (g)': 'mean'
})

#x_values are weight.  y_values are average tumor volume
x_values = average_volume_vs_weight['Weight (g)']
y_values = average_volume_vs_weight['Tumor Volume (mm3)']

#plotting weight vs average volume
plt.scatter(x_values, y_values)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen


#calculating standard linear regression variables
(slope,intercept,rvalue,pvalue,stderr) = st.linregress(x_values,y_values)

#calculating linear regression line
regress_values = x_values*slope + intercept
regress_values

#outputting linear correlation
print(f"The correlation between mouse weight and the average tumor volume is {round(rvalue,2)}")

#plotting scatterplot again with linear regression line
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
